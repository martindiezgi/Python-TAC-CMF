```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.by import By
from datetime import datetime
import os
import xlrd
import pandas as pd
```

Importamos packages a usar (selenium, datetime, os)
Luego definimos función a usar para scrap

```python
today = datetime.today().strftime('%Y%m%d') ## Descarga archivo CMF viene por default con el nombre costosffmm y la fecha de hoy
file_name = f"costosffmm{today}.xls" ## Por ello renombramos automaticamente de acuerdo a fecha del OS

#Función y scripts siguiente no son de mi autoría, genialidad de un amigo:

def month_year_iter( start_month, start_year, end_month, end_year ):
    ym_start= 12*start_year + start_month - 1
    ym_end= 12*end_year + end_month - 1
    for ym in range( ym_start, ym_end ):
        y, m = divmod( ym, 12 )
        yield y, m+1
```

Definiendo función de scrap (genialidad de alguien más, para variar):

```python
def scrap(yy, mm, b):
    b.get("https://www.cmfchile.cl/institucional/estadisticas/fmdfm_consulta1.php")

    select = Select(b.find_element(By.XPATH, '/html/body/div[2]/div[2]/div/div/div/div[4]/form/div[3]/select[1]'))
    select.select_by_value(m)

    select = Select(b.find_element(By.XPATH, '/html/body/div[2]/div[2]/div/div/div/div[4]/form/div[3]/select[2]'))
    select.select_by_value(y)

    b.find_element(By.XPATH, "/html/body/div[2]/div[2]/div/div/div/div[4]/form/div[4]/input").click()

    print(f"Descargando mes {m}, año {y}")
    while file_name not in os.listdir(r"C:/Users/Martin/Downloads/"): #Se pausa descarga hasta que encuentre nuevamente archivo en carpeta
        pass
    
    b.close
    print("Listo!")

b = webdriver.Chrome() #Especificar acá path del chromedriver o agregar a variables de entorno (yo hice lo 2do)

for yy, mm in month_year_iter(1, 2006, 6, 2024):
    y = f"{yy:04d}"
    m = f"{mm:02d}"
    scrap(y, m, b)
    desde = r"C:/Users/Martin/Downloads/" + file_name
    hacia = r"C:/Users/Martin/Desktop/Python/TAC/" + f"{y}{m}_TAC_CMF.xls" 
    os.rename(desde, hacia) #mandamos archivo desde carpeta default descargas Chrome a carpeta en que queremos los archivos
    print(f"{y}_{m} listo")

# Abajo de esto el status de la descarga
```

```python
# Define the directory path
path = ".\\TAC"

# List all files in the directory
tac_list = os.listdir(path)

# Filter for Excel files and create full paths
full_paths = [os.path.join(path, file) for file in os.listdir(path) 
              if file.endswith('.xls')]

# Initialize an empty list to store DataFrames
tac = []

# Loop through the list of full paths and read each Excel file into a DataFrame
for file_path in full_paths:
    df = xlrd.open_workbook(file_path, 
                            encoding_override='latin-1') #encoding genera problemas
    df = pd.read_excel(df)
    tac.append(df)

# Concatenate all DataFrames in the list into a single DataFrame
combined_df = pd.concat(tac, ignore_index=True)

# Save the combined DataFrame to a new CSV file
combined_df.to_csv("data_tac.csv", index=False)
```
